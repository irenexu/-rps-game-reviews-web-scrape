## Part I. Build Dataset
### Step 4 - Format & Save Dataset to CSV & folders

The result is store in csv file `game_reviews_rps_in_recent_years.csv` & txt files in game_reviews_rps_in_recent_years folder

*** the raw data (csv, json, txt) can all be found in test_output file, however, as 11 html code formats aren't enough to extract data accurately from all 2000+ reviews, there will be quite a lot rows with incorrect or missing data.

```python
import pandas as pd
import json
import os
```

```python
# Open the raw JSON file
with open('game_reviews_rps_in_recent_years.json', 'r') as file:
    # Load the JSON content into a variable
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
df
```

##### >>> Save Dataset to folder

```python
# save each review to 'game_reviews' directory
# original code obtained from class notebook - web-scrape-nursry-rhythms extract_lyrics.py
for i, element in enumerate(data):
    # Check if the 'game_reviews_rps' directory exist, and create it if needed
    if not os.path.exists(f'game_reviews_rps_in_recent_years'):
        os.makedirs(f'game_reviews_rps_in_recent_years')

    # Open a text file in the 'game_reviews' directory and write the reviews to it
    with open(f"game_reviews_rps_in_recent_years/{'_'.join(element['Title'].split())}.txt", "w") as f:
        f.write(element['Brief']) # add brief
        f.write("\n\n")
        f.write(element['Review']) # add main body text (review)
        f.close()
```

##### >>> Save dataset to csv

```python
# reference code found in: https://stackoverflow.com/questions/4622234/how-to-convert-a-list-to-a-csv-in-python
df.to_csv('game_reviews_rps_in_recent_years.csv', index=True)
```
