# NLP Project - Web Scraping and Text Analysis of Game Reviews on Rock Paper Shotgun
## Part I. Build Dataset
### Step 2 - Scrape webpages using bs4 & loops to get titles & links of all the reviews

The result is store in json file `rps_links_output_file`

```python
import requests
import json
from bs4 import BeautifulSoup
```

```python
scrapped_links = []

# the reviews section has 96 pages (until 1/12/2023) in total
for page in range(1, 97):
    RPS_LINK = f"https://www.rockpapershotgun.com/reviews?page={page}"
    response = requests.get(url=RPS_LINK)
    html_code = response.text
    soup = BeautifulSoup(html_code, "html.parser")

    titles = soup.select("p.title a.link.link--expand")

    for i in range(len(titles)):
        ds = {"Title": " ".join(titles[i].text.replace("\n","").split()),
              "URL": titles[i].get("href")
              }
        scrapped_links.append(ds)

print(scrapped_links)
```

```python
# reference code found in: https://stackoverflow.com/questions/25778021/how-can-i-save-a-list-of-dictionaries-to-a-file
# this will only need to be executed once! unless there is a need to update the dataset - latest update on 1/12/2023
with open('rps_links_output_file.json', 'w') as output:
    json.dump(scrapped_links, output)
```
